### Determinants using Laplace's Expansion

#### **Matrices:**

**Matrix A:**
```
2  3  1
1  4  0
3  2  1
```
**Determinant of A using Laplace's Expansion:**
```
2 × det(4 0
         2 1) - 3 × det(1 0
                           3 1) + 1 × det(1 4
                                          3 2)
= 2 × (4×1 - 0×2) - 3 × (1×1 - 0×3) + 1 × (1×2 - 4×3)
= 2 × 4 - 3 × 1 + 1 × (2 - 12)
= 8 - 3 - 10
= -5
```
