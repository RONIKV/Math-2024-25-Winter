### Matrix Operations in Python

We have the following matrices:

**Matrix A:**
```
1  2
3  4
```
**Matrix B:**
```
5  6
7  8
```
**Matrix C:**
```
-1  2
 3  0
```
**Matrix D:**
```
-1  2  3
 4  0  6
```
**Matrix E:**
```
1  2  4
5  7  8
```

#### Basic Operations
1. **A + B:**
```
6   8
10  12
```
2. **B - A:**
```
4  4
4  4
```
3. **A + C:**
```
 0  4
 6  4
```
4. **D + E:** _Addition not possible due to dimension mismatch._

#### Scalar Multiplication
1. **(1/2) * A:**
```
0.5  1.0
1.5  2.0
```
2. **2 * B:**
```
10  12
14  16
```
3. **-3 * C:**
```
 3  -6
-9   0
```
4. **4 * D:**
```
-4   8  12
16   0  24
```

#### Matrix Multiplication
1. **A * B:**
```
19  22
43  50
```
2. **B * A:**
```
23  34
31  46
```
3. **A * D:** _Multiplication not possible due to dimension mismatch._
4. **D * E:**
```
12  16  28
28  46  64
```
