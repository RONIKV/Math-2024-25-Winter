### Determinants of 2x2 and 3x3 Matrices

#### **2x2 Matrices:**

**Matrix A:**
```
2  3
1  4
```
**Determinant of A:**
```
(2×4) - (3×1) = 8 - 3 = 5
```

**Matrix B:**
```
5  6
7  8
```
**Determinant of B:**
```
(5×8) - (6×7) = 40 - 42 = -2
```

**Matrix C:**
```
-1  2
 3  0
```
**Determinant of C:**
```
(-1×0) - (2×3) = 0 - 6 = -6
```

#### **3x3 Matrices:**

**Matrix D:**
```
 1  0  2
-1  3  1
 2  4 -2
```
**Determinant of D:**
```
1×(3×-2 - 1×4) - 0×(-1×-2 - 1×2) + 2×(-1×4 - 3×2)
= 1×(-6 - 4) + 0×(2 - 2) + 2×(-4 - 6)
= (-10) + 0 + (-20)
= -30
```

**Matrix E:**
```
 3  1 -1
 0  2  4
 5  3  2
```
**Determinant of E:**
```
3×(2×2 - 4×3) - 1×(0×2 - 4×5) + (-1)×(0×3 - 2×5)
= 3×(4 - 12) - 1×(0 - 20) + (-1)×(0 - 10)
= 3×(-8) - (-20) - (-10)
= -24 + 20 + 10
= 6
```

**Matrix F:**
```
 2 -3  1
 1  4 -2
 1  5  3
```
**Determinant of F:**
```
2×(4×3 - (-2)×5) - (-3)×(1×3 - (-2)×1) + 1×(1×5 - 4×1)
= 2×(12 + 10) - (-3)×(3 + 2) + 1×(5 - 4)
= 2×22 + 3×5 + 1×1
= 44 + 15 + 1
= 60
```
