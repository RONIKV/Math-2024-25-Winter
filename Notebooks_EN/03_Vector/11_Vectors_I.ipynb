### Vectors I

#### **Scaling a Vector to Length 1**
Given vector:
```
a = [3, 4]
```
The length of vector \( a \) is calculated as:
```
|a| = sqrt(3² + 4²) = sqrt(9 + 16) = sqrt(25) = 5
```
To make its length equal to 1, we multiply it by:
```
1 / |a| = 1 / 5
```
So the unit vector of \( a \) is:
```
a' = [3/5, 4/5]
```

#### **Length and Unit Vector of Vector b**
Given vector:
```
b = [1, 1]
```
The length is:
```
|b| = sqrt(1² + 1²) = sqrt(1 + 1) = sqrt(2)
```
The unit vector of \( b \) is:
```
b' = [1/sqrt(2), 1/sqrt(2)]
```

#### **Plotting Vector and Unit Vector**
The vectors \( b \) and \( b' \) are plotted in a Cartesian coordinate system.

#### **Length and Unit Vector of Vector c**
Given vector:
```
c = [1, 2, 3]
```
The length is:
```
|c| = sqrt(1² + 2² + 3²) = sqrt(1 + 4 + 9) = sqrt(14)
```
The unit vector of \( c \) is:
```
c' = [1/sqrt(14), 2/sqrt(14), 3/sqrt(14)]
```

#### **Cartesian Coordinates of Vector v in Basis {b1, b2, b3}**
Given:
```
v = [2, 3, 4]
```
Basis vectors:
```
b1 = [1, 0, 1]
b2 = [0, 1, 0]
b3 = [1, 0, -1]
```
Solve the system:
```
v = a * b1 + b * b2 + c * b3
```
Expanding:
```
[2, 3, 4] = a * [1, 0, 1] + b * [0, 1, 0] + c * [1, 0, -1]
```
Setting up equations:
```
a + c = 2
b = 3
a - c = 4
```
Solving:
```
Adding the first and third equations:
( a + c ) + ( a - c ) = 2 + 4
2a = 6 → a = 3
```
Substituting \( a = 3 \) into \( a + c = 2 \):
```
3 + c = 2 → c = -1
```
Thus, the Cartesian coordinates of \( v \) in the given basis are:
```
(3, 3, -1)
```
