# Differential Equations

## First-Order Ordinary Differential Equations

### Problem 1:
Solve the first-order ordinary differential equations:

$$ \frac{dy}{dx} = y $$

#### Solution:
Using the **method of separation of variables**, we rewrite the equation:

$$ \frac{dy}{y} = dx $$

Integrating both sides:

$$ \int \frac{1}{y} dy = \int dx $$

$$ \ln |y| = x + C $$

Exponentiating both sides:

$$ y = Ce^x $$

where \( C \) is the constant of integration.

---

### Problem 2:
Solve:

$$ \frac{dy}{dx} = \frac{1}{2} y $$

#### Solution:
Rewriting:

$$ \frac{dy}{y} = \frac{1}{2} dx $$

Integrating:

$$ \ln |y| = \frac{x}{2} + C $$

Exponentiating:

$$ y = Ce^{x/2} $$

---

### Problem 3:
Solve:

$$ \frac{dy}{dx} = xy $$

#### Solution:
Rewriting:

$$ \frac{dy}{y} = x dx $$

Integrating:

$$ \ln |y| = \frac{x^2}{2} + C $$

Exponentiating:

$$ y = Ce^{x^2/2} $$

---

### Problem 4:
Solve:

$$ \frac{dy}{dx} = \frac{y}{x} $$

#### Solution:
Rewriting:

$$ \frac{dy}{y} = \frac{dx}{x} $$

Integrating:

$$ \ln |y| = \ln |x| + C $$

Exponentiating:

$$ y = Cx $$

## Second-Order Ordinary Differential Equations

### Problem 5:
Solve:

$$ y''(x) + y'(x) = 0 $$

with boundary conditions:

$$ y(0) = 2 \Rightarrow C_1 + C_2 = 2 $$
$$ y'(0) = -1 \Rightarrow -C_2 = -1 \Rightarrow C_2 = 1 $$

Thus, \( C_1 = 1 \), so:

$$ y(x) = 1 + e^{-x} $$

---

### Problem 6:
Solve:

$$ y''(x) - y(x) = 0 $$

with boundary conditions:

$$ y(0) = 2 \Rightarrow C_1 + C_2 = 2 $$
$$ y'(0) = 0 \Rightarrow C_1 - C_2 = 0 \Rightarrow C_1 = C_2 $$

Solving, \( C_1 = C_2 = 1 \), so:

$$ y(x) = e^x + e^{-x} $$

---

### Problem 7:
Solve:

$$ \frac{d^2 y}{dx^2} = -\omega^2 y $$

#### Solution:
The characteristic equation is:

$$ r^2 + \omega^2 = 0 $$

The roots are:

$$ r = \pm i\omega $$

Thus, the general solution is:

$$ y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x) $$

## Partial Differential Equations

### Problem 8:
Check if the function:

$$ \psi(t, x) = A \cos(\omega t + kx) $$

is a solution of the wave equation:

$$ \frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0 $$

where:

$$ v = \frac{\omega}{k} $$

#### Solution:
First, compute:

$$ \frac{\partial^2 \psi}{\partial t^2} $$

$$ \frac{\partial \psi}{\partial t} = - A \omega \sin(\omega t + kx) $$

$$ \frac{\partial^2 \psi}{\partial t^2} = - A \omega^2 \cos(\omega t + kx) $$

Now, compute:

$$ \frac{\partial^2 \psi}{\partial x^2} $$

$$ \frac{\partial \psi}{\partial x} = - A k \sin(\omega t + kx) $$

$$ \frac{\partial^2 \psi}{\partial x^2} = - A k^2 \cos(\omega t + kx) $$

Substituting into the wave equation:

$$ -A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx)) = 0 $$

Since:

$$ v^2 = \frac{\omega^2}{k^2} $$

It follows that the equation holds, confirming \( \psi(t, x) \) is a solution.
