### Equations of Lines on a Plane

#### **Equation of the Line through Points A(1,2) and B(3,4)**
Using the slope formula:
```
m = (y₂ - y₁) / (x₂ - x₁) = (4 - 2) / (3 - 1) = 2 / 2 = 1
```
Equation in point-slope form:
```
y - 2 = 1(x - 1)
```
Final equation:
```
y = x + 1
```

#### **Equation of the Line through A(1,2) Parallel to y = 2x + 3**
Since parallel lines have the same slope:
```
y - 2 = 2(x - 1)
```
Final equation:
```
y = 2x
```

#### **Equation of the Line through A(1,2) Perpendicular to y = 2x + 3**
Perpendicular slope is the negative reciprocal of 2:
```
m = -1/2
```
Equation:
```
y - 2 = (-1/2)(x - 1)
```
Final equation:
```
y = (-1/2)x + 5/2
```

#### **Intersection Point of y = 2x + 3 and y = 3x + 2**
Set the equations equal to each other:
```
2x + 3 = 3x + 2
```
Solve for x:
```
x = 1
```
Substituting into y = 2x + 3:
```
y = 5
```
Intersection point:
```
(1,5)
```

#### **Equation of Line through A(1,2) Parallel to Vector v = [2,3]**
The direction of the line is given by v:
```
m = 3/2
```
Equation:
```
y - 2 = (3/2)(x - 1)
```
Final equation:
```
y = (3/2)x + 1/2
```

#### **Example of a Line Perpendicular and Parallel to y = 2x + 3**
Parallel:
```
y = 2x + C
```
Example:
```
y = 2x - 4
```
Perpendicular:
```
y = (-1/2)x + C
```
Example:
```
y = (-1/2)x + 1
```

#### **Distance from Point A(1,2) to Line y = 2x + 3**
Using distance formula:
```
d = |Ax + By + C| / sqrt(A² + B²)
```
where A = 2, B = -1, C = -3, and point (1,2):
```
d = |2(1) - 1(2) - 3| / sqrt(2² + (-1)²)
  = |2 - 2 - 3| / sqrt(5)
  = | -3 | / sqrt(5)
  = 3 / sqrt(5)
```

#### **Equation of Line through A(2,0) and B(0,3)**
Slope:
```
m = (3 - 0) / (0 - 2) = -3/2
```
Equation:
```
y - 0 = (-3/2)(x - 2)
```
Final equation:
```
y = (-3/2)x + 3
```

#### **Angle between y = x + 3 and OX Axis**
Angle is given by:
```
theta = arctan(m)
```
where m = 1:
```
theta = arctan(1) = 45°
```
