### Equations of Planes in Space

#### **Equation of the Plane through Points A(1,2,3), B(3,4,5), and C(2,1,4)**
Find two direction vectors:
```
v1 = B - A = (3-1, 4-2, 5-3) = (2,2,2)
v2 = C - A = (2-1, 1-2, 4-3) = (1,-1,1)
```
Find normal vector as cross product:
```
n = v1 × v2 = |i  j  k|
               |2  2  2|
               |1 -1  1|
```
Solving determinant:
```
n = (2×1 - 2×-1, 2×1 - 2×1, 2×-1 - 2×1) = (2+2, 2-2, -2-2) = (4, 0, -4)
```
Equation of the plane:
```
4(x - 1) + 0(y - 2) - 4(z - 3) = 0
```
Simplify:
```
4x - 4z = -4 + 12
4x - 4z = 8 → x - z = 2
```

#### **Equation of the Plane through A(1,2,3) Parallel to 2x + 3y + 4z = 5**
Same normal vector:
```
2(x - 1) + 3(y - 2) + 4(z - 3) = 0
```
Simplify:
```
2x + 3y + 4z = 18
```

#### **Equation of the Plane through A(1,2,3) Perpendicular to Normal Vector n = [2,3,4]**
```
2(x - 1) + 3(y - 2) + 4(z - 3) = 0
```
Simplify:
```
2x + 3y + 4z = 18
```

#### **Intersection Line of Planes 2x + 3y + 4z = 5 and 3x + 4y + 2z = 6**
Solve system of equations:
```
Find direction vector by cross product of normals:
n1 = [2,3,4], n2 = [3,4,2]
n = n1 × n2 = (-10, 8, -1)
```
Find point on the line by solving equations with one variable set to 0.

#### **Equation of the Plane through A(1,2,3) Parallel to v1 = [1,0,1] and v2 = [0,1,-1]**
Find normal vector as cross product:
```
n = v1 × v2 = (-1, 1, 1)
```
Equation:
```
-1(x - 1) + 1(y - 2) + 1(z - 3) = 0
```
Simplify:
```
x - y - z = -2
```

#### **Example of Parallel and Perpendicular Planes to 2x + 3y + 4z = 5**
Parallel:
```
2x + 3y + 4z = 10
```
Perpendicular (normal vector different):
```
x + y - z = 3
```

#### **Distance from A(1,2,3) to Plane 2x + 3y + 4z = 5**
Formula:
```
d = |Ax + By + Cz + D| / sqrt(A² + B² + C²)
```
Substituting values:
```
d = |2(1) + 3(2) + 4(3) - 5| / sqrt(2² + 3² + 4²)
  = |2 + 6 + 12 - 5| / sqrt(4 + 9 + 16)
  = |15| / sqrt(29)
  = 15/sqrt(29)
```

#### **Equation of the Plane Intersecting Coordinate Axes at A(2,0,0), B(0,3,0), C(0,0,4)**
Using reciprocal coefficients method:
```
x/2 + y/3 + z/4 = 1
Multiply by 12:
6x + 4y + 3z = 12
```

#### **Angle Between Planes x + y + z = 1 and x = 0 (yz-plane)**
Angle formula:
```
cos(θ) = (n1 ⋅ n2) / (|n1| |n2|)
```
Where:
```
n1 = [1,1,1], n2 = [1,0,0]
cos(θ) = (1*1 + 1*0 + 1*0) / (sqrt(3) * 1)
       = 1 / sqrt(3)
θ = arccos(1/sqrt(3))
```
