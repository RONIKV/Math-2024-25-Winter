### Equations of Second-Order Curves (Conic Sections)

#### **Equation of a Circle**
Given center \( A(1,2) \) and radius \( r = 3 \), the equation is:
```
(x - 1)^2 + (y - 2)^2 = 9
```

#### **Equation of a Parabola**
Given x-intercepts at \( x = 2 \) and \( x = 4 \), the equation is:
```
y = a(x - 2)(x - 4)
```
Using the given point \( (3,1) \):
```
1 = a(3 - 2)(3 - 4)
1 = a(-1)
```
Solving for \( a \):
```
a = -1
```
Final equation:
```
y = -(x - 2)(x - 4)
```

#### **Finding the Center of an Ellipse**
Given equation:
```
x^2 + 4y^2 - 4x - 16y + 16 = 0
```
Rewriting:
```
(x^2 - 4x) + 4(y^2 - 4y) = -16
```
Completing the square:
```
(x - 2)^2 - 4 + 4[(y - 2)^2 - 4] = 0
(x - 2)^2 + 4(y - 2)^2 = 4
```
Center:
```
(2, 2)
```

#### **Finding the Slope of a Tangent Line to a Circle**
Given:
```
y = mx - 5
x^2 + y^2 = 1
```
Substituting \( y = mx - 5 \) into the circle equation and solving for \( m \) ensures tangency.

#### **Intersection of a Hyperbola and an Ellipse**
Given equations:
```
x^2 - y^2 = 1
x^2 + 4y^2 = 6
```
Solve the system to find intersection points.

#### **Distance Between Branches of a Hyperbola**
Given:
```
x^2 - y^2 = 1
```
The standard form is:
```
x^2/a^2 - y^2/b^2 = 1
```
Here, \( a^2 = 1 \) so \( a = 1 \).
Distance between branches:
```
2a = 2
```
