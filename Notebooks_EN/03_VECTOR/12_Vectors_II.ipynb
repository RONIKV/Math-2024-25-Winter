#### **Vector Addition and Plotting**

```
A = [2, 1]
B = [-1, 1]
```
**Addition:**
```
A + B = [2 + (-1), 1 + 1] = [1, 2]
```

#### **Area of Triangle Spanned by Two Vectors**
Given vectors:
```
A = [2, 1, 2]
B = [-1, 1, 1]
```
**Area Formula:**
```
Area = (1/2) * | A × B |
```
Compute cross product and find the magnitude to determine the area.

#### **Volume of Parallelepiped**
Given vectors:
```
A = [2, 1, -1]
B = [-1, 1, 0]
C = [1, 2, 1]
```
**Volume Formula:**
```
Volume = | A ⋅ (B × C) |
```
Compute cross product of B and C, then dot it with A to determine the volume.
